## 과제 2

```python
import pandas as pd
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
res = pd.merge(df1, df2, left_on="고객번호", right_on="고객번호", how="outer")
res
```

```python
# 과제2 : 금액에서 Nan 이라고 나온 부분을 0으로 채워보세요.
res['금액'] = res['금액'].fillna(0)
res
```

```python
# 과제 3. 아래와 같이 인구의 지역별 연도별 합계를 구하시오
group = df1.groupby(["지역", "연도"]).sum().unstack(["연도"])
group.reset_index()
group
```
