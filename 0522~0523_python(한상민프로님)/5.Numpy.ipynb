# 2. Numpy 배열 연산 
* 엑셀처럼 연산을 하게 하면서도 속도가 굉장히 빠르다
* numpy 연산의 핵심은 백터화!!
* numpy의 universial function을 통해 구현된다.
* 파이썬의 타입은 유연하다
* 이 사실은 효율적인 머신 코드로 컴파일 될 수 없다는 의미
* 연산을 할때마다 함수의 타입을 확인하고 그에 맞는 적절한 함수를 매칭하는데 많은 시간이 소요된다.

## 2.3 부울연산

```python
len(pd.read_csv('data/Seattle2014.csv')['PRCP'].values)
```

```python
rainfall = pd.read_csv('/content/content/MyDrive/content/Seattle2014.csv')['PRCP'].values
rainfall
```

```python
inches = rainfall/254
inches.shape
```

```python
# 1. 비가 온날의 일수는 몇 일?
비가온날 = 0
소규모강수량 = 0
for i in rainfall:
  비가온날 += (i>0)
  소규모강수량 += (i >0)&(i<=20)
print(비가온날)
print(소규모강수량)
```

```python
# 2. 최대로 비가 많이 온 날은 몇번째 일인가?
최대강수량 = max(rainfall)
최대강수일 = list(rainfall).index(최대강수량)
print(최대강수량)

# 3. 비가 온 날 중에서 강수량이 20 이하인 날은 몇일인가?
rainfall <= 20

```

```python
# 1. 비가 온날의 일수는 몇 일?
sum(rainfall>0)

# 2. 최대로 비가 많이 온 날은 몇번째 일인가?
idx = rainfall == np.max(rainfall)
rainfall[idx]

list(rainfall). index(np.max(rainfall))

# 3. 비가 온 날 중에서 강수량이 20 이하인 날은 몇 일인가?
idx = (rainfall > 0) & (rainfall <= 20)
sum(idx)
```

```python
index = (inches>0)&(inches<0.2)
np.min(inches[index])
```

```python
#부울연산자 활용
print('Number days without rain', np.sum(inches==0))
```

```python
print('number days with day', np.sum(inches>0))
print('Days with more than 0.5 inches', np.sum(inches>0.5))
print('Rainy days with < 0.1 inches', np.sum((inches>0) & (inches <0.2)))
```

```python
#마스킹 연산
#부울 배열을 마스크로 사용해 데이터의 특정 부분집합을 선택한다.
rainy = (inches > 0)
summer = (np.arange(365) - 172 <90) & (np.arange(365) - 172 >0)
print(np.median(inches[rainy])) # 비오는 날의 강수량 중앙값
print(np.median(inches[summer])) # 여름의 강수량 중앙값
print(np.max(inches[summer])) # 여름의 강수량 최대값
print(np.median(inches[rainy & ~summer])) # 여름제외 비오는 날의 강수량의 중앙값
```
