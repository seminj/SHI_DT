### 4-2. 확률 모형

- 무작위추출

- **확률**(probability)

- **확률 모형**(probability model)

- 주사위

#### 1) 확률의 기본

- 주사위 굴릴 때


- **확률변수**(random variable)
  - 취할 수 있는 값
  - 그 값이 나올 확률

- 주사위 굴리면

- **시행**(trial) / **실현값**(realization)

- **사건**(event) / **근원사건**(elementary event)

- 사건에 대응하여 확률 정의

- 확률의 수식 표현

$$
P(X=1) = \frac{1}{6}
\tag{4-1}
$$

- '눈이 홀수'인 사건

$$
\begin{align}
P\{(X=1) ∪ (X=3) ∪ (X=5)\}
&= P(X=1) + P(X=3) + P(X=5) \\\\
&= \frac{1}{6} + \frac{1}{6} + \frac{1}{6} \\\\
&= \frac{1}{2}
\end{align}
\tag{4-2}
$$

- 각각의 근원사건을 계산, 확률의 합을 계산
- 상호 배반이라면 그들의 합으로 계산할 수 있다. 

- 식(4-2)의 첫째 줄

- 배반사건의 확률의 합

- **상호배반**(mutually exclusive) : 서로 나올 수 없는 확률
- 단순히 더해서 구할 수없음.
 - **확률의 덧셈정리**:  상호배반이면 각각의 확률을 더해서 원하는 값을 구한다.
  - 두 사건 A, B가 배반사건이 아니라면 P(A∪B)=P(A)+P(B)-P(A∩B)이다.
  - 두 사건 A, B가 배반사건이면 P(A∪B)=P(A)+P(B)이다. 이 내용이 확률의 덧셈정리

- 예를 들어

- 상호배반이 아닌 예

- '눈이 홀수'인 확률 $=\frac{1}{2}$인 이유

#### 2) 확률분포

- 취할 수 있는 값과 그 값이 나오는 확률

```python
dice = [1, 2, 3, 4, 5, 6] 
prob = [1/21, 2/21, 3/21, 4/21, 5/21, 6/21]
# 확률분포, 눈과 확률 / 확률분포이구나를 알아야함
# 리스트로 각각 연결
```

- 확률변수 시행

```python
np.random.choice(dice, p=prob) # 랜덤값 생성
```

- 다음은 100번

```python
num_trial = 100 # 백번의 불공정한 시도
sample = np.random.choice(dice, num_trial, p=prob)
printobj(sample)
```

- 도수분포표 작성

```python
freq, _ = np.histogram(sample, bins=6, range=(1, 7))
pd.DataFrame({'frequency' : freq, # 도수분포
              'relative frequency' : freq / num_trial}, # 상대 도수분포
             index = pd.Index(range(1, 7), name='dice'))
```

- 히스토그램

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample, bins=6, range=(1, 7), density=True, rwidth=0.8)
# 실제의 확률분포를 가로선으로 표시
ax.hlines(prob, range(1, 7), range(2, 8), colors='gray')
# 막대 그래프의 [1.5, 2.5, ..., 6.5]에 눈금을 표시
ax.set_xticks(np.linspace(1.5, 6.5, 6))
# 주사위 눈의 값은 [1, 2, 3, 4, 5, 6]
ax.set_xticklabels(range(1, 7))
ax.set_xlabel('dice')
ax.set_ylabel('relative frequency')

plt.show()
print('[그림 4-2] 주사위 눈에 대한 히스토그램 (100번 시행)')
```

- 실제 확률분포와 상당한 차이

- 10000번 시행

```python
num_trial = 10000
sample = np.random.choice(dice, size=num_trial, p=prob)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample, bins=6, range=(1, 7), density=True, rwidth=0.8)
ax.hlines(prob, range(1, 7), range(2, 8), colors='gray')
ax.set_xticks(np.linspace(1.5, 6.5, 6))
ax.set_xticklabels(range(1, 7))
ax.set_xlabel('dice')
ax.set_ylabel('relative frequency')

plt.show()
print('[그림 4-3] 주사위 눈에 대한 히스토그램 (10000번 시행)')

# 백만번하면 신의 영역까지 도달한다 . 잘맞춤 /  모르는 정보를 일부의 정보로 맞춘다. 
```

```python
np.random.seed(0)
sample_means = [np.random.choice(scores, 20).mean() # 10,000회 중 20번뽑음
                for _ in range(10000)]

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample_means, bins=100, range=(0, 100), density=True)

# 모평균을 세로선으로 표시
ax.vlines(np.mean(scores), 0, 1, 'gray')
ax.set_xlim(50, 90)
ax.set_ylim(0, 0.13)
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')

plt.show()
print('[그림 4-7] 표본평균의 분포')
# 점수는 확률분포를 따르고 있다. 70.4 
# 개개인의 확률분포는 위에꺼를 따르지만  .
# 추론통계
# 중심극한의 정리 : 무작위 시행을 표본평균으로 하는 것. 개개의 점수가 아닌 표본평균을 알고자함. 그러기 위해 평균이 따르는 확률분포를 알아야함
```
