## 2.3 고급 인덱싱
* [행,열] 형식으로 추출할 수 있게 하여 가시성을 높여 준다.
* loc : 레이블(=키)을 기반으로 한다.
* iloc : 인덱스 값을 기반으로 한다.

```python
df = pd.DataFrame(np.arange(10, 22).reshape(3, 4),
                  index=["a", "b", "c"],
                  columns=["A", "B", "C", "D"])
df
```

### loc

```python
# 열선택
df.loc[:,'A']
```

```python
# 행선택
df.loc['b':'c']
```

```python
# 행선택 : A, C 만 가져오는 것
df.loc['b':'c', ['A','C']]
```

```python
#행,열선택
df.loc["b":, "A"]
```

```python
#부분 행,열 선택
df.loc[["a", "b"], ["B", "D"]]
```

```python
type(df.loc["b":,"A"]) #  Series(한개 열만 가져왔기 때문에) / 항상 type를 잘 체크할 것
type(df.loc["b",:])
```

```python
df.loc["b":,"A"].shape #Series
```

```python
type(df.loc["b":"b",:]) # DataFrame(여러 열 가져왔기 때문에)
```

```python
df.loc["b":"b",:].shape
```

```python
# 한개의 행 선택
df.loc["b", :]
```

```python
#조건부 행선택
df.loc[df.A>15]
```

```python
df
```

```python
# C > 12 이고 D가 16보다 큰 행들 추출
df.loc[df.C>12].loc[df.D>16]
# 이게 왜 될까?
```

```python
# C>12이고 D가 16보다 큰 행들을 추출하시오
idx = (df['C']>12)&(df['D']>16)
df.loc[idx,:]
```

```python
df.iloc[[False,True,True], [2,3]] # df.loc[df.A > 10, ["C", "D"]] # 숫자 대신에 TF로도 나타낼 수 있다. / 행열번호를 알고 있을 때 사용하는 것
```
