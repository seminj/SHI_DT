# 코랩 단축키 사용하기

## 셀추가 삭제
* ctrl + M + A : 앞에 셀 추가
* ctrl + M + B : 뒤에 셀 추가
* ctrl + M + D : 셀 삭제

## 셀타입 변경

* ctrl + M + Y : 코드 타입으로 변경
* ctrl + M + M : 마크다운으로 변경

## 실행

* ctrl + Enter : 실행하고 그대로 있기
* shit + Enter : 실행하고 커서는 다음셀로
* alt + Enter : 실행하고 셀삽입

## 셀 병합 / 분할
* 셀 병합 : (사용자 지정) -> Crtl + A
* 셀 분할 : (분기를 원하는 부분에 커서를 지정 후), Ctrl + M + -
